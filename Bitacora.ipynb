# 📓 INICIO DEL PROYECTO: “Spotify del vocabulario en inglés”

## 🧠 1. Idea general (resumen del proyecto)

Crear una plataforma web para aprender inglés basada en vocabulario personalizado, donde el usuario:

* Guarda palabras en “listas tipo playlists”
* Estudia esas listas en una sola pantalla
* Ve significado, pronunciación y ejemplos
* Escucha audio de pronunciación
* Aprende de forma visual, auditiva y por contexto

👉 Diferencia clave:
En lugar de cursos rígidos, el usuario estudia *solo las palabras que le interesan*, organizadas como música en Spotify.

## 🙍‍♂️ Casos de Usos

Se necesita responder a unas preguntas antes de empezar a dibujarel diagrama

### 📌 Regla principal del diagrama de casos de uso

> Un diagrama de casos de uso describe **qué puede hacer el usuario (o actor) con el sistema**, no cómo está implementado internamente.
>
> **Pregúntate siempre:**
>
> > "¿Esta acción la realiza o la solicita el usuario?"
>
> Si la respuesta es sí, probablemente es un caso de uso.
>
> Si la respuesta es no y describe algo interno del sistema, probablemente es una decisión técnica.

---

* Guardar palabra

❌ Decisiones técnicas

* Consultar una API
* Guardar en la base de datos
* Generar audio con Text-to-Speech
* Ejecutar una consulta SQL
* Cachear palabras

---

También te dejaría una versión aún más corta para ponerla como encabezado permanente:

> **Casos de uso = Qué hace el usuario.**
>
> **Arquitectura técnica = Cómo lo hace el sistema.**

Si durante el análisis te surge una idea y dudas dónde ponerla, prueba esta pregunta:

> **¿El usuario puede hacer clic en esto, solicitarlo o verlo como una función?**
>
> * Sí → Caso de uso.
> * No → Probablemente diseño técnico o arquitectura.

Esa regla te evitará el 90% de las confusiones mientras diseñas tu aplicación.

## Integración de imagenes, audio, 

### 👄 Imágenes de posición de boca

Tu idea sería algo así:

```text
wake up
↓
/w eɪ k ʌ p/
↓
[w] → imagen_01
[eɪ] → imagen_02
[k] → imagen_03
[ʌ] → imagen_04
[p] → imagen_05
```

Y luego la interfaz muestra:

👄 → 👄 → 👄 → 👄 → 👄

según los fonemas.

---

#### Ventaja enorme

No necesitas una imagen por palabra.

Necesitas una imagen por fonema.

Por ejemplo, el inglés tiene alrededor de:

* ~24 consonantes
* ~20 vocales

Dependiendo del sistema fonético.

En la práctica podrías tener:

* 30–50 imágenes

Y reutilizarlas para miles de palabras.

---

#### Espacio ocupado

Supongamos:

* 50 imágenes
* 100 KB cada una

Eso son:

```text
50 × 100 KB = 5 MB
```

Prácticamente nada.

---

### 🔊 Sobre el audio

Aquí sí hay varias opciones.

---

#### Opción 1: Grabar cada palabra

Ejemplo:

```text
wake-up.mp3
journey.mp3
effort.mp3
```

###### Problema

10,000 palabras.

Aunque cada audio pesara solo 20 KB:

```text
10,000 × 20 KB
≈ 200 MB
```

Y en la realidad probablemente sería más.

---

###### Ventajas

* Sonido natural
* Control total

###### Desventajas

* Muchísimo trabajo
* Mucho almacenamiento
* Difícil de mantener

---

#### Opción 2: Text-to-Speech (TTS)

Lo que usan muchas aplicaciones.

Tú guardas:

```text
wake up
```

Y cuando el usuario presiona reproducir:

un motor genera la voz.

---

###### Ventajas

* No almacenas audios
* Tamaño casi cero
* Fácil de escalar

---

###### Desventajas

* Dependes de un servicio

## Progreso de aprendizaje de palabras

El progreso de aprendizaje será global para cada usuario. Si una palabra se marca como aprendida, aparecerá como aprendida en cualquier lista y también en el buscador.

Por el momento el progreso de la palabra se mantiene dentro de la lista del usuario. Es decir si marca la palabra como aprendida dentro de su lista, solo aparecera aprendida en su lista, y si la ve en el buscador general aparecera normal (sin aprender)
